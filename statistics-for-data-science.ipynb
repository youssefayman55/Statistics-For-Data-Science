# What is statistics ? 
### Science of Statistics refers to the discipline that involves collecting, analyzing, interpreting, presenting, and organizing data. Statistics is essential for understanding trends, making decisions.

# why we need statistics ?
### 1- Data Interpretation : Statistics helps us interpret data effectively, turning raw data into meaningful insights. It identifies patterns and trends that help make informed decisions.

### 2- Decision Making : In business, government, healthcare, and many fields, statistics guide critical decisions by providing evidence-based analysis.

### 3- Predictive Analysis : It helps predict future trends and outcomes using historical data, assisting in areas like finance, marketing, and technology.


# Types Statistics : 
 ## Statistics is broadly divided into two main types:

## 1. Descriptive Statistics : Descriptive statistics summarizes and organizes data so it can be easily understood. This type of statistics does not draw conclusions or make inferences; it simply describes the data at hand.

## 2-  Inferential Statistics: Inferential statistics allows us to make predictions or inferences about a population based on a sample of data. This type of statistics goes beyond describing the data and helps in drawing conclusions. 

# 1. Descriptive Statistics

# what is population ?
### refers to All the people in a country and All the students in a school for a survey on study habits.

# what is samples ? 
### sample is a smaller, manageable subset of a population that is selected for the purpose of conducting a study or experiment. Researchers collect data from the sample to make inferences or draw conclusions about the entire population.

### 1- Representation 
### 2- Random

# what is variable ? 
###  variable is a characteristic, number, or quantity that can take different values in different individuals, objects, or events within a population or sample. Variables are used to capture information about the data being studied, and they are essential for performing statistical analysis.


# Types of Variables:

   ## These represent numeric values and can be measured. They are divided into two subcategories:
  ### 1 - Discrete Variables: Can only take specific, countable values (e.g., the number of students in a class).
  ### 2 - Continuous Variables: Can take any value within a given range (e.g., height, weight, temperature).
  
# 2- Qualitative (Categorical) Variables: 
  ## These represent categories or groups, and their values describe qualities or characteristics. They can be divided into:
  ### 1- Nominal Variables: Categories with no intrinsic order (e.g., gender, eye color, types of fruits).

# what is basic statistical measures ? 


## 1- Measures of Central Tendency : These measures describe the center or typical value in a dataset.
   ### * Mean (Average): The sum of all values divided by the number of values.
   ### * Median : The middle value when data is arranged in ascending or descending order. If the dataset has an even number of values, the median is the average of the two middle numbers.
  ### * Mode: The most frequently occurring value in a dataset.

  ## 2- Measures of Spread 
   ## * Range : The difference between the highest and lowest values in a dataset.
   #### Range=Maximum Value−Minimum Value
   
   ## * Variance : The average of the squared differences from the mean. It measures how much the data points deviate from the mean.
   ### Variance= ∑(Value−Mean) ** 2 / Number of Values

   ## * Standard Deviation : The square root of the variance. It shows the average distance between the data points and the mean.
   ### Standard Deviation=  square root of ( Variance )
   ## Note :  A small standard deviation means the data points are close to the mean; a large standard deviation indicates they are spread out.

# 3- Measures of Outliers 
### These measures describe the relative position of a data point within a dataset.
## Quartiles: Special percentiles that divide the data into four equal parts:

### => Q1 (First Quartile) : 25th percentile.
### => Q2 (Second Quartile) : 50th percentile (median).
### => Q3 (Third Quartile) : 75th percentile.
### => Interquartile Range (IQR) : The difference between the third quartile (Q3) and the first quartile (Q1). It represents the range of the middle 50% of the data.
 ### IQR=Q3−Q1 

### => Lower bound (minimum outlier threshold) : Lower bound = Q1 − 1.5 × IQR
### => Upper bound (maximum outlier threshold) : Upper bound = Q3 + 1.5 × IQR
### Note : Any data point below the lower bound or above the upper bound is considered an outlier.
## Example : 
### If 𝑄1 = 20 and 𝑄3 = 40 :
### IQR = 40 − 20 = 20
### Lower bound = 20 − 1.5 × 20 = 20 − 30 = −10
### Upper bound = 40 + 1.5 × 20 = 40 + 30 = 70 
### In this case, any value below −10 or above 70 would be considered an outlier.

# 4- Measures of Shape 
### These measures describe the distribution of data

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
import warnings
warnings.filterwarnings('ignore')
# Generating right-skewed data using an exponential distribution
np.random.seed(42)
data = np.random.exponential(scale=2, size=1000)  # Exponential distribution data
# Calculating skewness to confirm it's right-skewed
data_skewness = skew(data)
print(f"Skewness of the data: {data_skewness:.4f}")
# Plotting the right-skewed data using a histogram and density plot
plt.figure(figsize=(8, 6))
# Histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='black', color='lightgreen')
plt.title(f'Right-Skewed Data Histogram (Skewness: {data_skewness:.4f})')
plt.xlabel('Value')
plt.ylabel('Frequency')
# Density Plot using Seaborn
plt.subplot(1, 2, 2)
sns.kdeplot(data, shade=True, color="green")
plt.title(f'Right-Skewed Data Density Plot (Skewness: {data_skewness:.4f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
import warnings
warnings.filterwarnings('ignore')
# Generating left-skewed data by negating exponential distribution data
np.random.seed(42)
data = -np.random.exponential(scale=2, size=1000)  # Negated exponential distribution data
# Calculating skewness to confirm it's left-skewed
data_skewness = skew(data)
print(f"Skewness of the data: {data_skewness:.4f}")
# Plotting the left-skewed data using a histogram and density plot
plt.figure(figsize=(8, 6))
# Histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='black', color='lightcoral')
plt.title(f'Left-Skewed Data Histogram (Skewness: {data_skewness:.4f})')
plt.xlabel('Value')
plt.ylabel('Frequency')
# Density Plot using Seaborn
plt.subplot(1, 2, 2)
sns.kdeplot(data, shade=True, color="red")
plt.title(f'Left-Skewed Data Density Plot (Skewness: {data_skewness:.4f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
import warnings
warnings.filterwarnings('ignore')
# Generating symmetrical data using a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean = 0, Std = 1, 1000 data points
# Calculating skewness to confirm it's symmetrical
data_skewness = skew(data)
print(f"Skewness of the data: {data_skewness:.4f}")
# Plotting the symmetrical data using a histogram and density plot
plt.figure(figsize=(8, 6))
# Histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='black', color='lightblue')
plt.title(f'Symmetrical Data Histogram (Skewness: {data_skewness:.4f})')
plt.xlabel('Value')
plt.ylabel('Frequency')
# Density Plot using Seaborn
plt.subplot(1, 2, 2)
sns.kdeplot(data, shade=True, color="blue")
plt.title(f'Symmetrical Data Density Plot (Skewness: {data_skewness:.4f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

# Propability 
## Probability is a fundamental concept in statistics, mathematics, and everyday decision-making, representing the likelihood of an event occurring.

# 1- Basic Terminology (المصطلحات الاساسية) :
### 1- Experiment : An action or process that leads to one or more outcomes 
### 2- Outcome : The result of a single trial of an experiment (e.g., heads or tails in a coin toss).
### 3- Sample Space (S) : The set of all possible outcomes of an experiment.
  #### Example: For a die roll, the sample space is 𝑆={1,2,3,4,5,6} .
### 4- Event (E) : A specific set of outcomes from the sample space
  ### Example: In a coin toss, the event of getting a "head" is  E = { head }.

# 2. Probability of an Event (احتمال وقوع حدث ما) 
### The probability of an event 𝐴  occurring is calculated by the ratio of favorable outcomes to the total number of possible outcomes: 
   #### P(A) = Number outcomes / Totalnumber of outcomes
   #### Example: The probability of rolling a 3 on a fair six-sided die is: 
   ##### P(rolling a 3)= 1 / 6

# 3- Types of Event 
### 1. Simple Event : A simple event (or elementary event) is an event that consists of a single outcome.
 ### => Example: Rolling a 3 on a six-sided die is a simple event, as it involves only one specific outcome , {3}.

### 2. Compound Event: A compound event is an event that consists of two or more outcomes.
 ### => Example: Rolling an even number on a six-sided die is a compound event because it includes the outcomes {2,4,6} .

### 3. Mutually Exclusive Events (Disjoint Events) : Mutually exclusive events are events that cannot occur at the same time. If one event occurs, the other cannot.
### => Example: In a single roll of a die, the events of rolling a 2 and rolling a 5 are mutually exclusive. If you roll a 2, you cannot roll a 5 at the same time.

### 4. Independent Events : Two or more events are called independent events if the occurrence of one event does not affect the occurrence of the other.
### => Example: Rolling a die and flipping a coin are independent events. The result of the coin flip has no impact on the outcome of the die roll.

### 5. Dependent Events : Dependent events are events where the occurrence of one event affects the likelihood of the other.

# 4- complement : the complement of an event refers to the event not happening.  
### The probability of the complement of an event is calculated as: 
#### P(Ac) = 1 - P(A)
### Note : This means that the sum of the probabilities of an event and its complement must equal 1.

# 5- Conditional probability : refers to the probability of an event occurring given that another event has already occurred.
### The conditional probability of event 𝐴 given event B is written as P(A ∣ B) and is calculated using the formula:
#### => P(A|B) = P(A ∩ B) / p(B) 
#### => P(B | A) = P(A ∩ B) / p(A) 

# 6- Naive Bayes : 
### is a simple but powerful probabilistic machine learning algorithm based on Bayes' Theorem.  It is primarily used for classification tasks and is called "naive".  Naive Bayes performs well in many practical applications, especially in text classification problems like spam detection and sentiment analysis.
### Bayes' Theorem : The algorithm is based on Bayes' Theorem, which describes the probability of an event C (the class) given some evidence X (the features) as:
#### => P(C ∣ X) = P(X ∣ C) ⋅ P(C) / P(X)
### Naive Assumption :
### In Naive Bayes, the assumption is that all features in the data 𝑋 = {𝑥1,𝑥2,..,𝑥𝑛} are conditionally independent given the class C. This simplifies the calculation of P(X ∣ C) as:
#### => P(X ∣ C) = P(x1 ∣ C) . P(x2 ∣ C) . P(x3 ∣ C) ..... P(xn ∣ C)


# 2-  Inferential Statistics

# 1. Population vs. Sample : 
### => Population : The entire group of individuals or instances about which we want to draw conclusions.
### => Sample : A subset of the population selected for analysis. Inferential statistics uses the sample data to infer information about the larger population.
### Note : The goal is to ensure that the sample is representative of the population to make accurate inferences.

# 2. Parameter vs. Statistic 
### => Parameter : A numerical characteristic or measure that describes an aspect of the population (e.g., population mean 𝜇, population variance 𝜎2).
### => Statistic : A numerical characteristic or measure that describes an aspect of the sample (e.g., sample mean 𝑥ˉ, sample variance 𝑠2).

# 3. Point Estimation and Interval Estimation
### => Point Estimation : Provides a single value as an estimate of a population parameter (e.g., using the sample mean to estimate the population mean).
### => Interval Estimation (Confidence Intervals) : Provides a range of values within which the population parameter is expected to lie with a certain level of confidence (e.g., a 95% confidence interval for the population mean).
### Confidence Interval Formula for the Mean (when the population standard deviation is known):
#### CI = mean(sample) +- Z . 𝜎 / sqrt(n)
### => Z-value corresponding to the confidence level.
### => σ is the population standard deviation.
### => n is the sample size.

# 4. Hypothesis Testing
### Hypothesis testing is a formal method to make inferences or decisions about population parameters. It involves:
### Null Hypothesis (𝐻0) : A statement that assumes no effect or no difference (e.g., "The population mean is 5").
### Alternative Hypothesis (𝐻1 or 𝐻𝑎): A statement that contradicts the null hypothesis (e.g., "The population mean is not 5").
### The testing procedure includes:
### 1=> Set up hypotheses : Define 𝐻0 and 𝐻𝑎.
### 2=> Choose significance level (𝛼) : Common values are 0.05 or 0.01.
### 3=> Compute test statistic : Based on sample data (e.g., Z-test, t-test).
### 4=> Compare with critical value or p-value : Make a decision to reject or fail to reject 𝐻0.
### 5=> Conclusion : Draw an inference based on the test result.
### Common Hypothesis Tests:
### 1=> Z-Test : Used when the sample size is large (𝑛 ≥ 30) or the population standard deviation is known.
### 2=> T-Test : Used when the sample size is small (n < 30) and the population standard deviation is unknown.
### 3=> Chi-Square Test : Used to test the independence of two categorical variables.
### 4=> ANOVA (Analysis of Variance) : Used to compare the means of three or more groups to see if at least one group mean is different.

# 5. Types of Errors in Hypothesis Testing
### => Type I Error (False Positive) : Rejecting the null hypothesis when it is actually true.
### => Type II Error (False Negative) : Failing to reject the null hypothesis when it is actually false.

# 6. P-value
### The p-value is the probability of observing the sample data under the assumption that the null hypothesis is true. If the p-value is less than the chosen significance level (𝛼) , you reject the null hypothesis.

# 7. Confidence Levels and Significance Levels
### => Confidence Level : The percentage of times that the confidence interval would contain the population parameter if the same procedure were repeated (e.g., 95% confidence level).
### => Significance Level (𝛼) : The probability of rejecting the null hypothesis when it is true (e.g., 𝛼=0.05, meaning a 5% chance of making a Type I error).

# 8. Sampling Distributions
### A sampling distribution is the probability distribution of a given statistic (e.g., sample mean) based on a random sample. The most important concept here is the Central Limit Theorem (CLT):
### Central Limit Theorem: For a large enough sample size, the sampling distribution of the sample mean will be approximately normal, regardless of the distribution of the population. This is key to many inferential techniques, especially when dealing with sample means.
